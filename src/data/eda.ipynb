```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv("mock_interactions.csv")
```

```python
# count of each interaction type
interaction_counts = df['interaction_type'].value_counts()
print(interaction_counts)

# plot
sns.barplot(x=interaction_counts.index, y=interaction_counts.values)
plt.title("Distribution of Interaction Types")
plt.xlabel("Interaction Type")
plt.ylabel("Count")
plt.show()
```

```python
# number of unique items interacted with per user
unique_items_per_user = df.groupby('user_id')['item_id'].nunique()
print(unique_items_per_user.describe())  # summary stats

# plot histogram
plt.hist(unique_items_per_user, bins=10, edgecolor='black')
plt.title("Unique Items Interacted With per User")
plt.xlabel("Number of Unique Items")
plt.ylabel("Number of Users")
plt.show()
```

```python
# convert timestamp to datetime if not already
df['timestamp'] = pd.to_datetime(df['timestamp'])

# count interactions per day
df['date'] = df['timestamp'].dt.date
interactions_per_day = df.groupby('date').size()

# plot
plt.figure(figsize=(12,4))
interactions_per_day.plot(kind='line')
plt.title("Interactions Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Interactions")
plt.show()
```
